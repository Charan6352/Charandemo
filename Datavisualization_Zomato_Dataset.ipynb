# Matplotlib Practice with Zomato Data
Practising Data Visualization topic from class notes.


```python
from matplotlib import pyplot as plt
import numpy as np
import pandas as pd

df = pd.read_csv('zomato.csv')
df.head()
```

## Line plot ratings


```python
x = [1,2,3,4,5]
y = [2,3,4,3,5]
plt.plot(x,y)
plt.title('simple line')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```

## Scatter cost vs rating (small sample)


```python
sample = df[['avg cost (two people)','rate (out of 5)']].dropna().head(20)
x = sample['avg cost (two people)']
y = sample['rate (out of 5)']
plt.scatter(x,y)
plt.title('cost vs rating')
plt.xlabel('cost')
plt.ylabel('rating')
plt.show()
```

## Bar chart top areas


```python
areas = df['area'].value_counts().head(5)
x = areas.index
y = areas.values
plt.bar(x,y)
plt.title('top 5 areas')
plt.xlabel('area')
plt.ylabel('count')
plt.show()
```

## Horizontal bar chart restaurant type


```python
types = df['restaurant type'].value_counts().head(5)
x = types.values
y = types.index
plt.barh(y,x)
plt.title('top type')
plt.xlabel('count')
plt.ylabel('type')
plt.show()
```

## Histogram ratings


```python
x = df['rate (out of 5)'].dropna()
plt.hist(x)
plt.title('rating hist')
plt.xlabel('rating')
plt.ylabel('freq')
plt.show()
```

## Pie chart online order


```python
data = df['online_order'].value_counts()
x = data.values
labels = data.index
plt.pie(x,labels=labels)
plt.title('online order')
plt.show()
```

## Subplots example


```python
plt.figure(figsize=(8,6))
plt.subplot(2,2,1)
x = [1,2,3]
y = [2,4,1]
plt.plot(x,y)
plt.title('line')

plt.subplot(2,2,2)
x = areas.index
y = areas.values
plt.bar(x,y)
plt.title('areas')

plt.subplot(2,2,3)
plt.hist(df['rate (out of 5)'].dropna())
plt.title('hist')

plt.subplot(2,2,4)
plt.scatter(sample['avg cost (two people)'],sample['rate (out of 5)'])
plt.title('scatter')

plt.tight_layout()
plt.show()
```

```python
# heat map of correlation

data = df[['rate (out of 5)', 'num of ratings', 'avg cost (two people)']].dropna()
corr = data.corr()

plt.imshow(corr)
plt.colorbar()
plt.xticks(range(len(corr.columns)), corr.columns, rotation=45)
plt.yticks(range(len(corr.columns)), corr.columns)
plt.title('corr heat map')
plt.show()
```
